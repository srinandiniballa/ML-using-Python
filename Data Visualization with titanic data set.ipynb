```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
nn = pd.read_csv("titanic-training-data.csv")
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Pclass', data=nn)
plt.title('Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Sex', data=nn)
plt.title('By Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Embarked', data=nn)
plt.title('Embarked')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x='Pclass', y='Age', data=nn)
plt.title('Age distribution by class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass', y='Age', data=nn)
plt.title('Age distribution by class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='Pclass', y='Age', data=nn)
plt.title('Age distribution by class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
nn.hist(figsize=(10,10))
```

```python
nn.hist(figsize=(10,10), color='pink')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Pclass', hue='Sex', data=nn)
plt.title('Passengers by Class and Sex')
plt.xlabel('Class and Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Inference : Males are more in every class.
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Pclass', hue='Embarked', data=nn)
plt.title('Passengers by Class and Embarked')
plt.xlabel('Class and Embarked')
plt.ylabel('Count')
plt.show()
```

```python
# Inference : S are more in every class 
```

```python
corr = nn.corr()
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```
