### Outlier Treatment

```python
sns.boxplot(nn["Age"])
```

```python
q1=nn["Age"].quantile(0.25)
q3=nn["Age"].quantile(0.75)
iqr=q3-q1
```

```python
lower_threshold=q1-1.5*iqr
upper_threshold=q3+1.5*iqr
```

```python
lower_threshold
```

```python
upper_threshold
```

```python
nn=nn[(nn["Age"]>=lower_threshold)&(nn["Age"]<=upper_threshold)]
```

```python
sns.boxplot(nn["Age"])
plt.show()
```

```python
sns.boxplot(nn["SibSp"])
plt.show()
```

```python
q1=nn["SibSp"].quantile(0.25)
q3=nn["SibSp"].quantile(0.75)
iqr=q3-q1
lower_threshold=q1-1.5*iqr
upper_threshold=q3+1.5*iqr
```

```python
lower_threshold
```

```python
upper_threshold
```

```python
nn=nn[(nn["SibSp"]>=lower_threshold)&(nn["SibSp"]<=upper_threshold)]
```

```python
sns.boxplot(nn["SibSp"])
plt.show()
```

```python
median1=nn["Age"].median()
nn["Age"]=nn["Age"].fillna(median1)
mode1=nn["Embarked"].mode()[0]
nn["Embarked"]=nn["Embarked"].fillna(mode1)
nn.isnull().sum()
```

### Encoding

```python
nn=pd.get_dummies(nn,columns=["Sex","Embarked"])
```
